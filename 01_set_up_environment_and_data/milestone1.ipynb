```python

# number 8 on the milestone #1
import pandas as pd 
import numpy as np 



data = pd.read_csv('C:\\Users\\fderango\\Downloads\\01_set_up_environment_and_data\\us_stocks.csv')
data.shape
```

```python
# convert the date column  to datetime 

data['date'] = pd.to_datetime(data['date'])
```

```python
#Select stock price data only from 2,000 onwards
data = data.loc[30:,:]
data
```

```python
# reset the index with date and ticker  
data = data.set_index(['date','ticker'])
data
```

```python
# rename columns 
data.rename(columns={'adj_open':'open','adj_low':'low','adj_high':'high','adj_close':'close','adj_volume':'volumne'})
data
```

```python
# keep only certain columns
columns = ['open','low','high','close','volume']
new_df = data[columns]
new_df
```

```python

# number 8 on the milestone #1
import pandas as pd 
import numpy as np 



data = pd.read_csv('C:\\Users\\fderango\\source\\repos\\ml4trading\\01_set_up_environment_and_data\\us_stocks.csv')
data.shape
```

```python
data.head(5)
```

```python
# drop the nan's from the column 
data = data.dropna()
```

```python
# convert the date column  to datetime 

data['date'] = pd.to_datetime(data['date'])
```

```python
# self check 
data.info()
```

```python
#Select stock price data only from 2,000 onwards

#  # need to change this to find a better way 
# possibly use to date  

data = data.loc[data["date"] >= pd.to_datetime("2000-01-01")]
data
#data = data.loc[30:,:]
```

```python
# reset the index with date and ticker  
data = data.set_index(['ticker','date'])
data
```

```python
# only keeping certain columns such as all adjusted ones 'adj'
# keep only certain columns 
columns = ['adj_open','adj_low','adj_high','adj_close','adj_volume']
data = data[columns]
data
```

```python
# rename columns 
data = data.rename(columns={'adj_open':'open','adj_low':'low','adj_high':'high','adj_close':'close','adj_volume':'volumne'})
data
```

```python
# store the dataframe in HDF5 format

import h5py

data.to_hdf('data.h5', key='new_df', mode='w')
```
